```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Test3').getOrCreate()
print(spark)

df = spark.read.csv('employees_data.csv', header = True)
df.show()


df.createOrReplaceTempView('Employee')
result = spark.sql('''SELECT name,salary from Employee
                        WHERE department = "HR"''')
result.show(25)

bonus = result.withColumn('new_salary', result.salary*0.2+result.salary)
bonus.show()


avg_salary = spark.sql('''SELECT AVG(salary) as avg_salary,department
                            from Employee
                            Group By department''')
avg_salary.show()


count = spark.sql('''Select Count(name) as n,city
                    from Employee
                    group by city''')
count.show()

highest_paid = spark.sql('''Select name,salary
                            from Employee
                            Where department = "Sales"
                            order by salary desc
                            limit 1''')
highest_paid.show()

salary_per_dep = spark.sql('''Select Sum(Salary),department
                                from Employee
                                group by department''')
salary_per_dep.show()


youngest_emp = spark.sql('''Select name, age
                            from Employee
                            order by age asc
                            limit 5''')
youngest_emp.show()

high_salary_per_dep = spark.sql('''Select Sum(Salary),department
                                from Employee
                                group by department
                                limit 1''')
high_salary_per_dep.show()

names = spark.sql('''Select name
                    from Employee
                    where name like "A%"''')
names.show()
                  
```
